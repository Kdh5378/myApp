```python
import requests
from bs4 import BeautifulSoup
```

```python
headers={
    "user-agent":"Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36",
    "Accept-Encoding":"False"#오 이거 개쩐다
}

params={}#{"keyword":"한글"}

url="http://www.ppomppu.co.kr/zboard/zboard.php?id=freeboard"
```

```python
def getUrl(url, params=None, select="a", num_retrive=2):#구글 외의 사이트에서는 a만 찾아오게 한다
    html= requests.get(url, params=params, headers=headers)
    
    if 500<=html.status_code<600 and num_retrive>0:
        print(html.status_code, html.reason)
        return getUrl(url, params=params, num_retrive=num_retrive-1)
    
    dom= BeautifulSoup(html.text, "html.parser")
    links =dom.select(select)
    
    return [link["href"] for link in links if link.has_attr("href")==True]
```

```python
seed="http://www.ppomppu.co.kr/zboard/zboard.php?id=freeboard"
select="a"
queue=[seed]
```

```python
while queue:
    url = queue.pop()
    queue.extend(getUrl(url,params,select))#append는 list안의 list가 된다
    select="a"
    print(queue)
    break
```

#### 본문 긁어오기

```python
def getContent(url, params=None, select="a", num_retrive=2):#구글 외의 사이트에서는 a만 찾아오게 한다
    html= requests.get(url, params=params, headers=headers)
    
    if 500<=html.status_code<600 and num_retrive>0:
        print(html.status_code, html.reason)
        return getUrl(url, params=params, num_retrive=num_retrive-1)
    
    dom= BeautifulSoup(html.text, "html.parser")
    links =dom.select(select)
    
    return links
```

```python
seed="http://www.ppomppu.co.kr/zboard/zboard.php?id=freeboard"
params={"search_type":"sub_memo", "keyword":""}
contentList=getUrl(seed,params,"td.list_vspace img + a") # class가 list_vspace인 td의 모든 자식중 img다음에 있는 a
if params["keyword"]=="":
    contentList=contentList[1:]

seed="http://www.ppomppu.co.kr/zboard/"
for url in contentList:
    title=getContent(seed+url, params, ".view_title2")#>는 자기 첫째 자식 띄어쓰기는 자기 밑에 모든것 +는 형제중에 자기 다음꺼
    cont=getContent(seed+url, params, "table.pic_bg table td.han")
    comment=getContent(seed+url, params, "div.han")
    print("---------------------제목---------------------------")
    print(title[0].get_text())
    print("---------------------본문---------------------------")
    print(cont[0].get_text().strip()) #strip() 공백제거
    print("---------------------댓글---------------------------")
    for c in comment:  
        print(c.get_text().strip())
    print("----------------------끝---------------------------")
```
