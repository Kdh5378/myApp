```python
def getAirContent(url,params):
    sidoName=params["stationName"]
    params["serviceKey"]=urllib.parse.unquote(params["serviceKey"])
    params = urllib.parse.urlencode(params)
    params =params.encode("utf-8")
    #print(params)
    req=urllib.request.Request(url,data=params)
    res=urllib.request.urlopen(req)
    resResult=res.read()
    xml=etree.fromstring(resResult)
    xmlTree=etree.ElementTree(xml)
    root=xmlTree.getroot()
    items=root.findall(".//item")
    for item in items:
  #      print(item.text)#item당 20개의 요소들이 들어있음
        for i in range(0,20):
            print(item[i].text)#태그(키)와 value들
        conn.execute(sql, (sidoName,item[0].text, item[1].text, item[2].text, item[3].text, item[4].text, item[5].text, item[6].text, item[7].text, item[8].text, item[9].text, item[10].text, item[11].text, item[12].text, item[13].text, item[14].text, item[15].text, item[16].text, item[17].text, item[18].text, item[19].text))
        conn.commit()
```
