#### Hannanum

```python
from konlpy.tag import Hannanum
```

```python
h=Hannanum()
```

```python
text="분명히 <파 크라이 5>는 미국에서 광신도와 화약내 나는 친교를 즐기는 게임이었단 말이죠. 그런데 첫 번째 DLC '암흑의 시간'에서는 베트남 전쟁을 경험한 NPC의 과거를 다룬다는 이유로 주인공과 게임 배경을 바꿨습니다. 두 번째 DLC는 아예 화성으로 날아갑니다. 역시나 화성 이야기를 하는 NPC '닉 라이'가 있다는 이유입니다. <파 크라이 5: 화성에서 길을 잃다>는 화성에 납치된 닉 라이가 되어 외계인의 지구 침공 계획을 저지하는 내용을 담고 있습니다. 전투 중 외계 무기와 도구를 손에 넣어 게임을 풀고 지구로 무사 귀환하는 것이 목표입니다. 유비소프트의 <파 크라이 5: 화성에서 길을 잃다>는 7월 17일 정식 한국어화를 거쳐 출시되며, <파 크라이 5> 본편이 필요합니다. 세 번째 DLC '활보하는 좀비들'은 8월 말 출시 예정입니다."
```

```python
h.analyze(text)
```

```python
textSplit=text.split()
```

```python
candidates=h.analyze(text)
```

```python
for (i,t) in enumerate(textSplit):
    print(t)
    for c in candidates[i]:
        print (" ",c)
    print()
```

```python
print(h.pos(text), end="\n\n")
```

```python
print(h.morphs(text), end="\n\n")
```

#### Heap's law 새로운(중요한) 단어는 발생빈도수가 점점 줄어든다

```python
from konlpy.corpus import kolaw
from konlpy.tag import Kkma as kkma
from matplotlib import pyplot as plt
```

```python
k=kkma()
```

```python
text=kolaw.open(kolaw.fileids()[0]).read()
```

```python
nounLIst=k.nouns(text)
```

```python
nounList=nounLIst
```

```python
len(nounList)
global_unique=list()
global_unique_count=list()
```

```python
for row in text.split("\n"):
    term=k.nouns(row)
    unique=set(term)
    global_unique+=list(unique)
    global_unique =list(set(global_unique))
    #print(global_unique)
    global_unique_count.append(len(global_unique))
    print(len(unique), len(global_unique_count))
```

```python
plt.plot(global_unique_count)
```

```python
from konlpy.corpus import kobill
text=[kobill.open(i).read() for i in kobill.fileids()]
```

```python
len(text)
```

```python
for row in text:
    term=k.nouns(row)
    unique=set(term)
    global_unique+=list(unique)
    global_unique =list(set(global_unique))
    #print(global_unique)
    global_unique_count.append(len(global_unique))
    print(len(unique), len(global_unique))
```

```python
plt.plot(global_unique_count)
```
